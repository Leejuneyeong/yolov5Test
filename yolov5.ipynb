```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2 # 영상처리 cv2 -> opencv
```

```python
model = torch.hub.load('ultralytics/yolov5','yolov5s') # pytorch허브에서 기본 yolov5 모델을 가져옴
```

```python
model # sigmoid
```

```python
results = model(img_path)
results.print() # 실제로 해당 개체 감지 모델의 결과
```

```python
import json # json 파일 읽기
```

```python
with open('./data/tags_ko.json' ,encoding='UTF8') as f:
    tags_ko = json.load(f)

tags_ko
```

```python
# 이미지 사진을 넣고 yolov5로 돌리고 이미지 분석하고 데이터베이스에 라벨값 넣기
# glob 함수 : 사용자가 제시한 조건에 맞는 파일명을 리스트 형식으로 반환
from glob import glob

img_list = glob('static/photos/*.jpg')

# 딕셔너리 
db={}

for img_path in img_list:
    results = model(img_path)
    
    tags = set() # 중복을 없애주는 내장함수 set()
    
    for pred in results.pred[0]:
        tag = tags_ko[int(pred[-1])] # 각 이미지의 좌표값, 라벨값에서 맨 마지막이 라벨값임 맨 마지막값만 추출하고 
        # json 파일의 한국어로 번역하게 바꿔줌 당연히 형변환 해줘야함
        tag = tag.replace(' ','')
        tags.add(tag)
    db[img_path] = list(tags)
    
db
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html', photos=db)

if __name__ == '__main__':
    app.run()
```
